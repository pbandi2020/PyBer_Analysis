```python
%matplotlib inline

# import dependencies
import matplotlib.pyplot as plt
# matplotlib.__version__
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Line chart using object oriented programming
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# the above 2 line can also me written as follows and get same results
# fig = plt.figure() 
# ax = fig.add_subplot()
# ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create a plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis,marker="*", color="blue", linewidth=1, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# showing xlim Jan thry June
# plt.xlim("Jan","June")
plt.xlim(0,5)
# ax.xlim("Jan","June)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
# figure height and width
plt.figure(figsize=(20,10))
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal chart
# plt.barh(y_axis, x_axis)
plt.barh(x_axis, y_axis,color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend() 
         
         
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - horizontal
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
# plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend() 
```

```python
# scattered plot. Both command below gives the same result
# plt.plot(x_axis, y_axis, 'o')
plt.scatter(x_axis, y_axis, color="red", label='Chicago')
plt.gca().invert_xaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend() 
```

```python
# managing the size if the plot, Default is using the "s=y_axis", Below is more a customized.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)

# plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# getting the same results us list comprehenssion logic

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#using object oriented method. Basically same results using diffrent line of code.

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)

```

```python
# plt.pie(y_axis, labels=x_axis)
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
